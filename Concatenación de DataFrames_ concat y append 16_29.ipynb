```python
df1 = pd.DataFrame(np.random.rand(3,2)*10,columns=['a','b'],index=[0,1,2])
df1
```

```python
df2 = pd.DataFrame(np.random.randn(3,2)*-1,columns=['a','b'],index=[2,3,4])
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner')  #funciona como sql, en este caso para la intersecciÃ³n
```

```python
pd.concat([
           df1.reset_index(drop=True),
           df2.reset_index(drop=True)
           ],axis=1)  #para unificar indices
```
