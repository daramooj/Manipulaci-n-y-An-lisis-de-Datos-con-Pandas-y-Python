```python
import pandas as pd
pd.__version__
```

```python
data = {'names':['Sara Moreno 34',
                 'jUAn GOMez 23',
                 'CArlos mArtinez 89',
                 'Alfredo VelaZques 3',
                 'luis Mora 56',
                 '@freddier #platzi 10',pd.NA]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['names'].str.lower() #hace que todas las letra se conviertan a minúsculas
```

```python
df['names'].str.upper() #hace que todas las letra se conviertan a mayúsculas
```

```python
df['names'].str.capitalize() #solo la primera letra en mayúscula
```

```python
  df['names'].str.len() #Para contar la longitud de nuestro texto usamos
```

```python
df['names'].str.split('#')  #Para dividir el texto por espacios usamos split y definimos el carácter por el que queremos dividir, en este caso, un espacio vacío ' ' o '#'
```

```python
df['names'].str[:10] #Para seleccionar los primeros o últimos 5 o cualquier otro número de caracteres, usamos
```

```python
df['names'].str[-5:]
```

```python
df['names'].str.replace('Alfredo','Antonio') #Podemos reemplazar una secuencia de caracteres por otra mediante
```

```python
df['names'].str.findall('ara') #podemos buscar una secuencia de texto en específico, en este caso,'ara'
```

```python
df['names'].str.contains('or')  #También podemos crear un filtro basándonos en una secuencia de texto en específico, en este caso, las filas que tengan 'or'
```

```python
df['names'].str.lower().str.count('a') #Así mismo, podemos contar el número de ocurrencias de un caracter en específico,por ejemplo, cuántas veces aparece la letra 'a':
```

```python
df['names'].str.extract('([0-9]+)', expand=False) #usando Regex, para extraer los caracteres numéricos
```

```python
df['names'].str.replace('@[^\s]+','') #si quiero extraer las menciones '@xxxx' del texto:
```
