```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('diamonds')
df
```

```python
df.groupby('cut').mean() #para obetener el promedio
```

```python
df.groupby('cut').median() #un estimador estadistico de media (?)
```

```python
df.groupby('cut')['carat'].median() #aquí agrupamos por cut en la columna carat y sacamos la mediana (o valor medio de los datos)
```

```python
df.groupby('cut')['carat'].count() #aquí agrupamos por cut en la columna carat y contamos los datos
```

```python
df.groupby('cut')['carat'].max() #aquí agrupamos por cut en la columna carat y obtenemos el valor máximo
```

```python
df.groupby('cut')['carat'].max() #aquí agrupamos por cut en la columna carat y obtenemos el valor mínimo
```

```python
for i in range (0,4):
  print(i)
```

```python
for key_group, group in df.groupby('cut'):
  grouped_price = group['price'].mean()

print('Cut: {}, Price: {} \n'.format(key_group,grouped_price))
```

```python
df.groupby(['cut','color'])['price'].mean().to_frame
```

```python
df.groupby(['cut','color'])['price'].agg(['min',np.mean,max])
```

```python
def mean_kilo(x):
  return np.mean(x)/1000
```

```python
df.groupby(['cut','color'])['price'].agg(['min',np.mean,max,mean_kilo]).head(10) #esto me muestra que con agg puedo agregar funciones creadas poor mi a mi groupby
```

```python
dict_agg = {'carat':[min,max], 'price':[np.mean, mean_kilo]}
dict_agg
```

```python
df.groupby(['cut','color']).agg(dict_agg) #este es un Ej. de que se puede agregar un diccionario a groupby 
```

```python
def f_filter(x):
  return mean_kilo(x['price'])>4
```

```python
df.groupby(['cut','color']).filter(f_filter) #esto indica que podemos definir filtros y con ellos agregarlos al groupby
```
