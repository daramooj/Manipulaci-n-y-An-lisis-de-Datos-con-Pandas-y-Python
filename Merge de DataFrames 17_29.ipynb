```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
    'Mix':['y2','y3','a2','a3']},
    index = [0,1,2,3])
df_left
```

```python
df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])
df_right
```

```python
pd.merge(df_left,df_right,how='inner',on='Y')
```

```python
pd.merge(df_left,df_right,how='inner',left_on='Mix', right_on='Y', suffixes=['_izq','_der']) #suffixes cambia los nombre de los sufijos uando las columnas que comparten los df tienen el mismo nombre sino se lo damos ella los coloca por defecto
```

```python
pd.merge(df_left,df_right,how='inner',left_on='Mix', right_on='A')
```

```python
df_left1 = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3']},
    index= [0,1,2,3])
df_left1
```

```python
pd.merge(df_left1,df_right,how='outer',on='Y') #si comparte más de una columna que queremos intersectar en on colocamos una lista por ejemplo las columnas "A" y "Y" entonces on=['A','Y']
```
