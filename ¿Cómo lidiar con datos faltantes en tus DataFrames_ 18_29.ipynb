```python
import numpy as np
import pandas as pd
```

```python
pd.options.mode.use_inf_as_na = True #La versión 1.0 de pandas incluye un nuevo objeto NA, que es mucho más
                                     #general pues, ademas de interactuar con números, tambien puede hacerlo con
                                     #cadenas de texto u otras varaibles como las de tipo booleano.
```

```python
df = pd.DataFrame(np.arange(0, 15).reshape(5, 3), columns=['a', 'b', 'c'])
df
```

```python
df['d'] = np.nan
df['e'] = np.arange(15, 20)
df.loc[5,:] = pd.NA
df.loc[4,'a'] = pd.NA
df.loc[0,'d'] = 1
df.loc[5,'d'] = 10
df
```

```python
df.isnull() #Para reconocer cuando un objeto es nulo 
```

```python
df.isna() #Para reconocer cuando un objeto es nulo 
```

```python
df.isnull().sum() #Conocer el número de variables nulas por columna
```

```python
df.notnull().sum(axis=1) #el número de filas con elementos no nulos, por ejemplo en la fila 0 hay 5 elementos no nulos
```

```python
df.size-df.isnull().sum().sum()   #todos los elementos no nulos de nuestro DataFrame
```

```python
df[df['a'].notnull()] #filtrar por las variables no nulas de la columna ‘a’
```

```python
df.dropna() #dropna es perfecto para elimnar rapidamente las filas con registros faltantes
```

```python
df[['a']].dropna()
```

```python
df.fillna(0) # Usando la función fillna podremos reemplazarlas por el valor que querramos, en este caso 0
```

```python
df.fillna(method="ffill") #Si quisieramos remplazar con el valor siguiente
```

```python
df.fillna(method="bfill") #Si quisieramos remplazar con el valor previo 
```

```python
df.fillna(method="bfill",axis=1) #El mismo ejercicio anterior se puede aplicar con las filas 
```

```python
fill = pd.Series([100, 101, 102]) #Podemos usar también una serie para reemplazar los valores de una columna en especifico, es importante que haya emparejamiento entre los índices
fill
```

```python
df['d'] = df['d'].fillna(fill)
df['d'] 
```

```python
df
```

```python
 df.fillna(df.median()) #Una de las formas más usadas para reemplazar datos es usar el promedio de las columnas, esto se hace con la función mean. O si se quiere un mejor estimador, usamos median.
```
